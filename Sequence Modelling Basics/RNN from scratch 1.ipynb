```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
data_train = pd.read_csv("/kaggle/input/amazon-reviews/train.csv")
data_train = data_train.sample(500000, random_state=99)
data_test = pd.read_csv("/kaggle/input/amazon-reviews/test.csv")
```

```python
data_train.columns = ['polarity','title','text']
data_test.columns = ['polarity','title','text']
```

```python
data_train['text'] = data_train['title']+ data_train['text']
data_test['text'] = data_test['title']+ data_test['text']
data_train.head()
```

```python
data_train = data_train.drop('title',axis=1)
data_test = data_test.drop('title',axis=1)
```

```python
def clean_text(text):
    if pd.isnull(text):
        return ""
    text = re.sub(r"<.*?>", "", text)
    text = re.sub(r"[^a-zA-Z]", " ", text)
    return text.lower()
```

```python
data_train['text'] = data_train['text'].astype(str).apply(clean_text)
data_train.head()
```

```python
data_test['text'] = data_test['text'].astype(str).apply(clean_text)
data_test.head()
```

```python
from collections import defaultdict
vocab = defaultdict(lambda: len(vocab))
UNK = vocab["<UNK>"]
def tokenize(text):
    MAX_LEN = 100
    tokens = [vocab[word] for word in text.split()[:MAX_LEN]]
    padding_needed = MAX_LEN - len(tokens)
    if padding_needed > 0:
        tokens += [vocab["<PAD>"]] * padding_needed
    return tokens

PAD = vocab["<PAD>"]

tokenized = data_train['text'].apply(tokenize)
tokenized2 = data_test['text'].apply(tokenize)
```

```python
vocab_size = len(vocab)
vocab_size
```

```python
max_len = 100
X_train = np.array([np.pad(x,(0, max_len - len(x)),constant_values=0) for x in tokenized])
y_train = data_train['polarity'].values
X_test = np.array([np.pad(x,(0, max_len - len(x)),constant_values=0) for x in tokenized2])
y_test = data_train['polarity'].values
X_train = np.array(X_train).astype(np.int32)
X_test = np.array(X_test).astype(np.int32)
```

```python
class SimpleRNN:
    def __init__(self, vocab_size, embedding_dim, hidden_dim, max_len):
        self.embedding_dim = embedding_dim
        self.hidden_dim = hidden_dim
        self.max_len = max_len

        self.E = np.random.randn(vocab_size, embedding_dim).astype(np.float32) * 0.01
        self.Wx = np.random.randn(embedding_dim, hidden_dim).astype(np.float32) * 0.01
        self.Wh = np.random.randn(hidden_dim, hidden_dim).astype(np.float32) * 0.01
        self.bh = np.zeros((hidden_dim,), dtype=np.float32)
        self.Why = np.random.randn(hidden_dim, 1).astype(np.float32) * 0.01
        self.by = np.zeros((1,), dtype=np.float32)

    @staticmethod
    def sigmoid(x):
        x = np.clip(x, -500, 500)
        return 1 / (1 + np.exp(-x))

    
    def forward(self, x_batch):
        batch_size = x_batch.shape[0]
        x_batch = x_batch.astype(np.int32)
        self.h_states = np.zeros((self.max_len, batch_size, self.hidden_dim), dtype=np.float32)
        h = np.zeros((batch_size, self.hidden_dim), dtype=np.float32)

        for t in range(self.max_len):
            x_t = self.E[x_batch[:, t]]
            h = np.tanh(np.dot(x_t, self.Wx) + np.dot(h, self.Wh) + self.bh)
            self.h_states[t] = h

        self.final_h = h
        self.out = self.sigmoid(np.dot(h, self.Why) + self.by)
        return self.out

    
    def compute_loss(self, pred, label):
        label = label.reshape(-1, 1).astype(np.float32)
        return -np.mean(label * np.log(pred + 1e-8) + (1 - label) * np.log(1 - pred + 1e-8))

    
    def backward(self, x_batch, y_batch, lr=0.01):
        batch_size = x_batch.shape[0]
        x_batch = x_batch.astype(np.int32)
        dy = self.out - y_batch.reshape(-1, 1)

        dWhy = np.dot(self.final_h.T, dy)
        dby = np.sum(dy, axis=0)

        dh = np.dot(dy, self.Why.T)

        dWx = np.zeros_like(self.Wx)
        dWh = np.zeros_like(self.Wh)
        dbh = np.zeros_like(self.bh)
        dE = np.zeros_like(self.E)

        for t in reversed(range(self.max_len)):
            h = self.h_states[t]
            h_prev = self.h_states[t - 1] if t > 0 else np.zeros_like(h)
            x_t = self.E[x_batch[:, t]]

            dh_raw = dh * (1 - h ** 2)
            dWx += np.dot(x_t.T, dh_raw)
            dWh += np.dot(h_prev.T, dh_raw)
            dbh += np.sum(dh_raw, axis=0)
            np.add.at(dE, x_batch[:, t], np.dot(dh_raw, self.Wx.T))
        
            dh = np.dot(dh_raw, self.Wh.T)

        self.Why -= lr * dWhy
        self.by -= lr * dby
        self.Wx -= lr * dWx
        self.Wh -= lr * dWh
        self.bh -= lr * dbh
        self.E -= lr * dE
```

```python
def train_model(model, X_train, y_train, epochs=10, lr=0.01, batch_size=64):
    n = len(X_train)
    for epoch in range(epochs):
        permutation = np.random.permutation(n)
        X_train = X_train[permutation]
        y_train = y_train[permutation]
        correct = 0
        total_loss = 0

        for i in range(0, n, batch_size):
            X_batch = X_train[i:i + batch_size]
            y_batch = y_train[i:i + batch_size]

            preds = model.forward(X_batch)
            loss = model.compute_loss(preds, y_batch)
            total_loss += loss

            correct += np.sum((preds > 0.5).astype(int).flatten() == y_batch)

            model.backward(X_batch, y_batch, lr=lr)

        acc = correct / n
        print(f"Epoch {epoch+1} - Loss: {total_loss:.4f} - Accuracy: {acc*100:.2f}%")
```

```python
def evaluate_model(model, X_test, y_test, batch_size=64):
    correct = 0
    for i in range(0, len(X_test), batch_size):
        X_batch = X_test[i:i + batch_size]
        y_batch = y_test[i:i + batch_size]
        pred = model.forward(X_batch)
        correct += np.sum((pred > 0.5).flatten().astype(int) == y_batch)
    return correct / len(X_test)
```

```python
vocab_size = len(vocab)
embedding_dim = 32
hidden_dim = 64
max_len = 100

model = SimpleRNN(vocab_size, embedding_dim, hidden_dim, max_len)
train_model(model, X_train, y_train, epochs=2, lr=0.01, batch_size=64)
test_acc = evaluate_model(model, X_test, y_test)
print(f"Test Accuracy: {test_acc*100:.2f}%")
```
