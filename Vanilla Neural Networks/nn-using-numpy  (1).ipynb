```python
data =  np.array(data)
m,n = data.shape
np.random.shuffle(data)
data.shape
```

```python
data_dev = data[0:14000].T
Y_dev = data_dev[n-1]
X_dev = data_dev[0:n-1]

data_train = data[14000:m].T
Y_train = data_train[n-1]
X_train = data_train[0:n-1]
```

```python
def RELU(Z):
    return np.maximum(0, Z)

def RELU_deriv(Z):
    return Z > 0

def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))
    
def compute_loss(A2, Y,w0 =1, w1 = 4):
    m = Y.shape[0]
    return -np.sum( w1 * Y * np.log(A2 + 1e-12) + w0 * (1 - Y) * np.log(1 - A2 + 1e-12)) / m
    
def init_params():
    W1 = np.random.randn(64, 10) * 0.01
    b1 = np.zeros((64, 1))
    W2 = np.random.randn(1,64) * 0.01
    b2 = np.zeros((1, 1))
    return W1, b1, W2, b2
    
def forward_prop(W1, b1, W2, b2, X):
    Z1 = W1.dot(X) + b1
    A1 = RELU(Z1) 
    Z2 = W2.dot(A1) + b2
    A2 = sigmoid(Z2)
    return Z1, A1, Z2, A2
def backward_prop(Z1, A1, A2, W2, X, Y):
    m = X.shape[0]
    dZ2 = A2 - Y
    dW2 = (1 / m) * dZ2.dot(A1.T)
    db2 = (1 / m) * np.sum(dZ2, axis=1, keepdims=True)
    
    dZ1 = (W2.T.dot(dZ2)) * RELU_deriv(Z1) 
    dW1 = (1 / m) * dZ1.dot(X.T)
    db1 = (1 / m) * np.sum(dZ1, axis=1, keepdims=True)
    
    return dW1, db1, dW2, db2
def update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha):
    W1 -= alpha * dW1
    b1 -= alpha * db1
    W2 -= alpha * dW2
    b2 -= alpha * db2
    return W1, b1, W2, b2
    
```

```python
def get_predictions(A2):
    return A2 > 0.25

def get_accuracy(predictions, Y):
    return np.mean(predictions == Y)

def gradient_descent(X, Y, alpha, iterations):
    W1, b1, W2, b2 = init_params()
    
    for i in range(iterations):
        Z1, A1, Z2, A2 = forward_prop(W1, b1, W2, b2, X)
        loss = compute_loss(A2, Y)
        dW1, db1, dW2, db2 = backward_prop(Z1, A1, A2, W2, X, Y)
        W1, b1, W2, b2 = update_params(W1, b1, W2, b2, dW1, db1, dW2, db2, alpha)
        
        if i % 50 == 0 or i == iterations - 1:
           print("Iteration: ", i)
           print("A2 values:\n", A2[:10])
           print("Accuracy: ", get_accuracy(get_predictions(A2), Y))
    
    return W1, b1, W2, b2
```

```python
W1, b1, W2, b2 = gradient_descent(X_train, Y_train, alpha=0.00001, iterations=500)
```

```python
def predict(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_prop(W1, b1, W2, b2, X)
    return A2 > 0.3
    
```

```python
testing = predict(X_dev,W1,b1,W2,b2)
predictions = testing.astype(int)
```
